```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import Birch
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data
data = pd.read_csv('Mall_Customers.csv')
```

```python
# Select features (let's assume 'Annual Income (k$)' and 'Spending Score (1-100)')
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
'''Definnition: BIRCH (Balanced Iterative Reducing and Clustering using Hierarchies) clustering is a hierarchical clustering algorithm 
designed for handling large datasets efficiently.
It aims to partition data points into a hierarchical structure of clusters, which can be efficiently navigated and analyzed.'''
# Apply BIRCH clustering
birch = Birch(threshold=0.5, n_clusters=5)
cluster_labels = birch.fit_predict(X_scaled)
```

```python

# Plotting the clusters
plt.figure(figsize=(8, 6))
#sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=data, hue=cluster_labels, palette='viridis')
# Using a sequential palette
#sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=data, hue=cluster_labels, palette='plasma')
# Using a diverging palette
#sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=data, hue=cluster_labels, palette='coolwarm')
# Using a custom palette
#custom_palette = ["#FF5733", "#33FF57", "#3357FF"]  # Example custom palette with three colors
#sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=data, hue=cluster_labels, palette=custom_palette)
#sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=data, hue=cluster_labels, palette='viridis', hue_norm=(0, 10))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=data, hue=cluster_labels, palette='coolwarm', style=cluster_labels)


plt.title('BIRCH Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```

```python
# Select features (let's assume 'Age', 'Annual Income (k$)', and 'Spending Score (1-100)')
X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply BIRCH clustering
birch = Birch(threshold=0.5, n_clusters=5)
cluster_labels = birch.fit_predict(X_scaled)
```

```python
# Plotting the clusters in 3D
fig = plt.figure(figsize=(14, 12))
ax = fig.add_subplot(111, projection='3d')

# Plotting data points with cluster labels and different styles
for cluster_label in np.unique(cluster_labels):
    ax.scatter(X.loc[cluster_labels == cluster_label, 'Age'],
               X.loc[cluster_labels == cluster_label, 'Annual Income (k$)'],
               X.loc[cluster_labels == cluster_label, 'Spending Score (1-100)'],
               label=f'Cluster {cluster_label}',
               cmap='viridis',
               marker=f'${cluster_label}$')  # Use LaTeX-style markers for different cluster labels

# Labeling the axes
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
ax.set_title('BIRCH Clustering')
ax.legend()

plt.show()
```
